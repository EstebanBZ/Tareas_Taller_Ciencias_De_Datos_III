```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
ruta = "./for007.dat"

# Leer archivo
data = pd.read_csv(ruta, delim_whitespace=True, header=None)

#print(data.head())
```

```python
# Asigna las columnas a variables
col_1 = data.iloc[:, 0]  # Columna 1
col_2 = data.iloc[:, 1]  # Columna 2 
col_3 = data.iloc[:, 2]  # Columna 3
col_4 = data.iloc[:, 3]  # Columna 4
col_6 = data.iloc[:, 16]  # Columna 6
col_9 = data.iloc[:, 8]  # Columna 9
log_col_1 = np.log10(data.iloc[:, 0] - 1.0)
log_col_6 = np.log10(data.iloc[:, 5])
```

```python
# Crea figura de 4x2
fig, axs = plt.subplots(nrows=4, ncols=2, figsize=(16, 21))

# Primer grfico en el primer subplot
axs[0, 0].plot(col_2, col_3, label="HYDWIND", color="blue")
axs[0, 0].set_xlim([-1.0, 0])
axs[0, 0].set_ylim([0, 450])
axs[0, 0].set_xlabel('Columna 2')
axs[0, 0].set_ylabel('Columna 3')
axs[0, 0].set_title('Velocity Profile')
axs[0, 0].grid(True)
axs[0, 0].legend(loc='upper left')

# Segundo grfico en el segundo subplot
axs[0, 1].plot(col_1, col_3, color="blue")
axs[0, 1].set_xlim([1.000, 1.040])
axs[0, 1].set_ylim([0.0, 20.0])
axs[0, 1].set_xlabel('Columna 1')
axs[0, 1].set_ylabel('Columna 3')
axs[0, 1].set_title('V(r/R*) (inner)')
axs[0, 1].grid(True)

# Tercer grfico en el tercer subplot
axs[1, 0].plot(col_2, col_3, label='HYDWIND', color="blue")
axs[1, 0].plot(col_2, col_6, label='TRIAL PROFILE', color="red")
axs[1, 0].set_xlim([-1.0, 0])
axs[1, 0].set_ylim([0, 450])
axs[1, 0].set_xlabel('Columna 2')
axs[1, 0].set_ylabel('Columna 3 / Columna 17')
axs[1, 0].set_title('Velocity Profile')
axs[1, 0].grid(True)
axs[1, 0].legend(loc='upper left')

# Cuarto grfico en el cuarto subplot
axs[1, 1].plot(col_1, col_3, color='blue')
axs[1, 1].set_xlim([70, 100])
axs[1, 1].set_ylim([380, 420])
axs[1, 1].set_xlabel('Columna 1')
axs[1, 1].set_ylabel('Columna 3')
axs[1, 1].set_title('V(r/R*) (outer)')
axs[1, 1].grid(True)

# Quinto grfico en el quinto subplot
axs[2, 0].plot(col_2, col_4, color='blue')
axs[2, 0].set_xlim([-1.0, 0])
axs[2, 0].set_ylim([0, 139])
axs[2, 0].set_xlabel('Columna 2')
axs[2, 0].set_ylabel('Columna 4')
axs[2, 0].set_title('dw/du')
axs[2, 0].grid(True)

# Sexto grfico en el sexto subplot
axs[2, 1].plot(log_col_1, log_col_6, color='blue')
axs[2, 1].set_xlim([-4.0, 2.1])
axs[2, 1].set_ylim([-18.1, -9.4])
axs[2, 1].set_xlabel('log (Columna 1 - 1.0)')
axs[2, 1].set_ylabel('log (Columna 6)')
axs[2, 1].set_title('Density')
axs[2, 1].grid(True)

# Septimo grfico en el septimo subplot
axs[3, 0].plot(col_1, col_9, color='blue')
axs[3, 0].set_xlim([-1.0, 20])
axs[3, 0].set_ylim([-200, 1600])
axs[3, 0].set_xlabel('Columna 1')
axs[3, 0].set_ylabel('Columna 9')
axs[3, 0].set_title('g_line')
axs[3, 0].grid(True)

# Elimina el octavo subplot vaco
fig.delaxes(axs[3, 1])

# Agrego un Titulo general
fig.suptitle('Subplots Esteban Barrios', fontsize=22, fontweight='bold')

# Ajusta los espacios entre los subplots
fig.tight_layout()

plt.show()
```

```python
# Asignar columnas a variables
col_2 = data.iloc[:, 1]  # Columna 2
col_3 = data.iloc[:, 2]  # Columna 3

# Crea una figura y un eje
fig, ax = plt.subplots()

# Grafica la lnea
ax.plot(col_2, col_3, label = "HYDWIND", color = "blue")

# Establece lmites en los ejes
ax.set_xlim([-1.0, 0])
ax.set_ylim([0, 450])


# Etiqueta los ejes
ax.set_xlabel('Columna 2')
ax.set_ylabel('Columna 3')
ax.set_title('Velocity Profile')

# Agrega las grids
ax.grid(True)

# Hacer que se grafique la etiqueta de la linea
ax.legend(loc='upper left')

plt.show()
```
