Importar las siguientes librer√≠as: `matplotlib.pyplot, seaborn, pandas, numpy`

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

## <b>Recrea la siguiente figura</b>

```python
# Generar valores de x en el rango deseado
x = np.linspace(-4, 4, 52)

# Calcular valores para curvas
y1 = 3 * np.exp(-0.25 * x**2)
y2 = 2.8 * np.exp(-0.15 * (x - 0.1)**2)

# Graficar curvas
plt.plot(x, y1, color='red', linestyle='-.', linewidth=2, label='3e^(-0.25x^2)')
plt.plot(x, y2, color='cyan', marker='o', markerfacecolor='yellow', markeredgecolor='black', markersize=6, linestyle='-', linewidth=2, label='2.8e^(-0.15(x-0.1)^2)')

# Configurar etiquetas y t√≠tulo del gr√°fico
plt.xlabel("That took a wile", fontweight='bold')
plt.title('Many Plots', fontweight='bold')
plt.legend(loc='upper right')

# Crear gr√°fico m√°s peque√±o en el centro
# Coordenadas del gr√°fico central
plt.axes([0.41, 0.18, 0.2, 0.3])
plt.text(-1.5, 2, "Cool", color='red')
plt.text(-0.5, 0, "Plot", color='blue')
plt.text(1, -2, "Bro", color='green')

plt.title('Small Plot')
plt.xlim(-2, 2)
plt.ylim(-3, 3)
plt.xticks(np.arange(-2, 2.5, 1))
plt.yticks(np.arange(-3, 3.5, 1))

plt.show()
```

## <b> Recrea el siguiente `jointplot` </b>

## <b> Recrea los siguientes `boxplots` </b>

```python
# Graficar boxplots
plt.figure(figsize=(8.4, 4.2))
sns.boxplot(data=abnb, x='neighbourhood_group', y='price')
plt.xlabel('Neighbourhood Group')
plt.show()
```

```python
#No ejecutar este m√≥dulo hasta que est√©s content@ con tu resultado. Al ejecutar el m√≥dulo el gr√°fico desaparecer√°.
```

```python
plt.figure(figsize=(8.4, 4.2))

sns.set(style="ticks")

# Graficar boxplots
sns.boxplot(data=abnb, x='neighbourhood_group', y='number_of_reviews', hue='room_type')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Number of Reviews')
plt.ylim(-20, 370)

# Mostrar el gr√°fico
plt.show()
```

## <b> Recrea el siguiente  `violinplot` comparando la distribuci√≥n de  `Entire home/apt` y `Private room` para los 5 `neighbourhood groups`</b>

```python
# Filtrar los datos para las categor√≠as de inter√©s (Entire home/apt y Private room)
filtered_abnb = abnb.loc[abnb['room_type'].isin(['Entire home/apt', 'Private room'])]
sns.set(style="ticks")

# Ajustar tama√±o de la figura
plt.figure(figsize=(10, 6))
# Graficar violinplot
sns.violinplot(data=filtered_abnb, x='neighbourhood_group', y='price', hue='room_type', split=True)

# Establecer etiquetas
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')

plt.show()
```


## <b>Time Series: Recrea el siguiente gr√°fico</b>

(Pista: Convierte la columna `last_review` a formato `DateTime`)

```python
# Convertir la columna "last-review" al formato DateTime
abnb['last_review'] = pd.to_datetime(abnb['last_review'])
abnb_timepo = abnb[(abnb['last_review'] >= pd.to_datetime('2016-10-01')) & (abnb['last_review'] < pd.to_datetime('2019-03-01'))]

sns.set(style="ticks")

# Ajustar tama√±o de la figura
plt.figure(figsize=(10, 6))

sns.lineplot(data=abnb_timepo, x='last_review', y='number_of_reviews', label='Number of Reviews')
sns.lineplot(data=abnb_timepo, x='last_review', y='price', label='Price')

plt.xlabel('last_review')
plt.ylabel('')
plt.title('Fluctutations in Number of Reviews and Price over time')

plt.legend()
plt.show()
```
