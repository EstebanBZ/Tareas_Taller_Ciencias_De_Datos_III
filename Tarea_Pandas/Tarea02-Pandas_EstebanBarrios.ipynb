>   **no** sean `female` o `male`. Asigna cada valor de `sex` en el nuevo `DataFrame` a
>   un nuevo valor de 'x'. Determina el número total de valores `null` en el subconjunto.
>
> 2. Crea un nuevo objeto `DataFrame` que contenga solo observaciones cuyos valores en la columna `sex` sean `male`
>   o `female` y en los cuales el valor de `weight` sea mayor que 0. 

```python
# Seleccionar filas sex con diferente F y M
df = surveys_df[(surveys_df['sex'] != 'female') & (surveys_df['sex'] != 'male')].copy()

# Asignar X a cada valor
df['sex'] = 'x'

# Contar el número total de valores nulos
revisando_uno = df.isnull().sum().sum()

print(revisando_uno)
```

```python
# Seleccionar todas las filas con sex igual a 'M' o 'F' y weight mayor que 0
revisando_uno = surveys_df[(surveys_df['sex'].isin(['M', 'F'])) & (surveys_df['weight'] > 0)]

revisando_uno
```

 # Pandas 04 **Combinando DataFrames con Pandas**

> ## Distribuciones
>
> Cree un nuevo DataFrame uniendo los contenidos de survey.csv y species.csv. Luego calcula:
>
>   - taxa por parcela
>   - taxa por sexo por parcela


```python
species_df = pd.read_csv('species.csv')
```

```python
#species_df.head()
```

```python
# Une los DataFrames en uno
df_combinado = pd.merge(surveys_df, species_df, on='species_id')

df_combinado
```

```python
# Número de especies de cada taxa por parcela
taxa_parcela = df_combinado.groupby(['plot_id', 'taxa'])['species_id'].nunique()

# Imprimir el resultado
taxa_parcela
```

```python
# Número de especies de cada taxa por sexo en cada parcela
taxa_sexo_parcela = df_combinado.groupby(['plot_id', 'taxa', 'sex'])['species_id'].nunique()

taxa_sexo_parcela
```
